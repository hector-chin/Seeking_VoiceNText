```python
from ckiptagger import construct_dictionary, WS, POS
ws = WS("./data")
pos = POS("./data")
```

```python
import json
import os
```

```python
for file in os.listdir('./youtube'):
    if file.endswith('_醫療保健text.json'):
        print(file)
```

```python
with open('./stopword.txt', encoding='utf-8') as f:
    stop = f.read()
```

```python
stopword = stop.split('\n')
stopword
```

```python
alltext = []
for x in os.listdir('./podcast'):
    if x.endswith('_醫療保健text.json'):
        with open(f'./podcast/{x}') as f:
            data = json.loads(f.read())
        for file in data.values():
            for name, texts in file.items():
                word_sentence_list = ws(list(texts))
                pos_sentence_list = pos(word_sentence_list)
                for i in range(len(texts)):
                    for word in zip(word_sentence_list[i], pos_sentence_list[i]):
                        alltext.append(word)
```

```python
alltext = list(set(alltext))
```

```python
for i in alltext:
    if i[1] not in ['A','Na','Nb','Nc','Nd','Nv','VA','VAC','VB','VC','VD','VF','VE','VG','VH','VHC','VI','VJ','VL','FW']:
        if i[0] not in stopword:
            stopword.append(i[0])
            with open('stopword.txt', 'a', encoding='utf-8') as f:
                f.write('\n'+i[0])
```
