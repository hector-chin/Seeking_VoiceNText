```python
from pydub import AudioSegment
import speech_recognition as sr
import os
import time
import json
```

```python
if not os.path.exists('./分割暫存'):
    os.mkdir('./分割暫存')
```

```python
channels = ['./youtube/' ,'./podcast/']
r = sr.Recognizer()
for channel in channels:
    chname = channel.split('/')[-2]

    print(chname)
    file_names = [fn for fn in os.listdir(channel) if os.path.isdir(channel+fn)]
    for fn in file_names:
        texts = {}
        print(fn)
        cf = channel+fn
        m4a_names = [mn for mn in os.listdir(cf) if mn.endswith('.mp4')]
        for mn in m4a_names:
            texts.update({mn: []})
            print(mn)
            title = mn[:-4]
            path = cf + '/' + title
            try:
                episode = AudioSegment.from_file(path+'.mp4')
                for x in range(int(len(episode)/(30*1000))+1):
                    epath = './分割暫存/{}.wav'.format(x)
                    episode[(x*30)*1000:((x+1)*30+0.2)*1000].export(epath, format='wav')
                    with sr.WavFile(epath) as source:
                        audio = r.record(source)
                    try:
                        start = time.time()
                        text = r.recognize_google(audio,language='zh-tw')
                        end = time.time()
                        texts[mn].append(text)
                        print(text)
                        print(end - start)
                    except:
                        print('Error')
            except:
                pass
            print('====================episode end====================')
        print('====================channel end====================')
        with open('./{}/{}_text.json'.format(chname, fn), 'w') as f:
            json.dump(texts,f)
    print('====================category end====================')
```
