```python
import pandas as pd
data = pd.read_csv('traindata')
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(data.iloc[:,:-1],data.iloc[:,-1],test_size=0.3,random_state=0)
```

```python
from keras.models import Sequential
from keras.layers.core import Dense, Activation
```

```python
inputdim = X_train.shape[1]
```

```python
model = Sequential()
model.add(Dense(512, input_dim=inputdim))
model.add(Activation('relu'))
model.add(Dense(256))
model.add(Activation('relu'))
model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dense(4))
model.add(Activation('softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x=X_train,y=y_train, validation_split=0.1, epochs=10, verbose=1)  
```

```python
model.evaluate(X_test, y_test ,batch_size=10)
```

```python
X_testttt = X_test.to_numpy()
print(X_testttt)
```

```python
model.predict(X_testttt)
```
