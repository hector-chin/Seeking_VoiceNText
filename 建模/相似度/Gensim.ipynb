```python
import json
with open('./youtube/醫療保健text.json') as f:
    doc_json = json.loads(f.read())
```

```python
doc_tokenized = []
for i in doc_json.values():
    for x, y in i.items():
        word_sentence_list = ws(list(y))
        doc_list = []
        for z in word_sentence_list:
            doc_list.extend(z)
        doc_tokenized.append(set(doc_list))
        break
    break
```

```python
import gensim
import pprint
from gensim import corpora
from gensim import models
dictionary = corpora.Dictionary()
BoW_corpus = [dictionary.doc2bow(doc, allow_update=True) for doc in doc_tokenized] 
import numpy as np
# fit tfidf model (ntc=raw tf+zero corrected idf+cosine normalization)
tfidf = models.TfidfModel(BoW_corpus, smartirs='ntc')
```

```python
from gensim import similarities
index = similarities.SparseMatrixSimilarity(tfidf[BoW_corpus], num_features=12)
```

```python
with open('./podcast/哇賽心理學_醫療保健text.json') as f:
    query_json = json.loads(f.read())
query_document = []
for k in query_json.values():
    for l in k.values():
        query_list = ws(list(l))
        for c in query_list:
            query_document.extend(c)
        break
    break
query_bow = dictionary.doc2bow(list(set(query_document)))
sims = index[tfidf[query_bow]]
```

```python
for document_number, score in sorted(enumerate(sims), key=lambda x: x[1], reverse=True):
    print(document_number, score)
```
