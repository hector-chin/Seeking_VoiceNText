```python
import json
import os
```

```python
lists = ['./youtube/政治.json']
```

```python
with open('./標題分類2.json') as f:
    data_json = json.loads(f.read())
category = list(data_json.keys())
for i in lists:
    with open(i) as f:
        datas = json.loads(f.read())
    for data in datas.values():
        for eps, url in data.items():
            if eps[:-4] in [y for x in data_json.values() for y in x.keys()]:
                continue
            print(eps)
            cat = input('輸入類別:')
            if cat in category:
                data_json[cat].update({eps[:-4]: url})
            else:
                data_json[cat] = {eps[:-4]: url}
                category.append(cat)
```

```python
with open('./標題分類final.json','w') as f:
    json.dump(data_json,f)
```
