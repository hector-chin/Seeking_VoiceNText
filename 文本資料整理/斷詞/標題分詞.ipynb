```python
from ckiptagger import data_utils, construct_dictionary, WS, POS, NER
import json
import os
```

```python
ws = WS("./data")
```

```python
lists = ['./podcast/醫療保健.json']
titlelist = []
for x in lists:
    with open(x) as f:
        y = json.loads(f.read())
    for z in y.values():
        for i in z.keys():
            titlelist.append(i)
```

```python
with open('newworddict.txt', encoding='utf-8') as f:
    newwordlist = f.read()
word_to_weight = {}
newwordlist = newwordlist.split('\n')
for word in newwordlist:
    word_to_weight.update({word: 1})
dictionary = construct_dictionary(word_to_weight)
```

```python
word_sentence_list_new = ws(list(titlelist), recommend_dictionary=dictionary)
```

```python
for i, sentence in enumerate(titlelist):
    print(sentence)
    for word in word_sentence_list_new[i]:
        print(word, end='//')
    print()
    newword = input('新增詞彙(不用n):')
    while newword != 'n':
        if newword in newwordlist:
            pass
        else:
            with open('newworddict.txt', 'a', encoding='utf-8') as f:
                f.write(newword+'\n')
            newwordlist.append(newword)
        newword = input('新增詞彙(不用n):')
```
