```python
import jieba
import json
```

```python
with open('./youtube/醫療保健text.json') as f:
    data = json.loads(f.read())
```

```python
#官方支援繁體的jieba辭典,檔案放置於當前資料夾
jieba.set_dictionary('jiebadict.txt')
#github上有人整理出台灣的用詞
jieba.load_userdict('jiebadict_TW.txt')
```

```python
#提出文本
for file in data.values():
    for name, texts in file.items():
        print(name)
        for text in texts:
            print(text)
################################# jieba斷詞 #################################
            seg_list0 = jieba.cut(text, cut_all=True)#全模式
            seg_list1 = jieba.cut(text)#精準模式
            seg_list2 = jieba.cut_for_search(text)#搜尋模式
            print('全模式:')
            print("/ ".join(seg_list0))
            print('精準模式:')
            print("/ ".join(seg_list1))
            print('搜尋模式:')
            print("/ ".join(seg_list2))
            result = jieba.tokenize(text)
#             result = jieba.tokenize(text, mode='search')
            for tk in result:
                print(tk)
###############################################################################
            break
        break
    break
```
