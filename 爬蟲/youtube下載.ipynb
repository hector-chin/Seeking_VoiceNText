```python
from pytube import Channel, YouTube
from datetime import datetime
import re
import os
import json
```

```python
def validateTitle(title):
    rstr = r"[\/\\\:\*\?\"\<\>\|]"  # '/ \ : * ? " < > |'
    new_title = re.sub(rstr, "_", title)  # 替換為下劃線
    return new_title
```

```python
def youtubedownload(channellist, url_json, afterdate):
    youtube = []
    for channel in channellist:
        ytchannel = Channel(channel)
        path = './youtube/{}/'.format(validateTitle(ytchannel.channel_name))
        if not os.path.exists(path):
            os.makedirs(path)
        print(ytchannel.channel_name)
        for url in ytchannel.video_urls:
            yt = YouTube(url)
            date = yt.publish_date
            if date >= datetime.strptime(afterdate, "%Y%m%d"):
                filename = validateTitle(yt.title) + '.mp4'
                if os.path.exists(path+'/'+filename):
                    youtube.append({'Title': validateTitle(yt.title), 'Url': url, 'Date': date.strftime("%Y%m%d")})
                    print(filename)
                else:
                    try:
                        yt.streams.get_audio_only().download(output_path=path, filename=filename)
                        print(filename+'...'+'Done')
                        youtube.append({'Title': validateTitle(yt.title), 'Url': url, 'Date': date.strftime("%Y%m%d")})
                    except:
                        print('===========An Error Occurred===========')
            else:
                break
        print('=='*30)
    with open(f'./youtube/{url_json}.json', 'w') as f:
        json.dump(youtube,f)
    print('===========All Finished===========')
    return
```
