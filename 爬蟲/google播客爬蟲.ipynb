```python
def validateTitle(title):
    rstr = r"[\/\\\:\*\?\"\<\>\|]"  # '/ \ : * ? " < > |'
    new_title = re.sub(rstr, "_", title)  # 替換為下劃線
    return new_title
```

```python
def googlepodcast(channellist, afterdate, url_json):
    podcast = []
    for name in channellist:
        url = 'https://podcasts.google.com/search/{}'.format(name)

        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.54 Safari/537.36'
        }
        res = requests.get(url, headers=headers)

        soup = BeautifulSoup(res.text, 'html.parser')

        url2 = 'https://podcasts.google.com' + soup.select('a[class="yXo2Qc"]')[0]['href'][1:]

        res2 = requests.get(url2, headers=headers)

        soup2 = BeautifulSoup(res2.text, 'html.parser')

        ChannelName = soup.select('a[class="yXo2Qc"]')[0].select('div[class="FyxyKd"]')[0].text
        print(ChannelName)
        print()
        for index, episode in enumerate(soup2.select('div[class="oD3fme"]')):
            episodeurl = 'https://podcasts.google.com' + soup2.select('a[class="D9uPgd"]')[index]['href'][1:]
            date = episode.select('div[class="OTz6ee"]')[0].text
            if 'ago' in date:
                if 'hour' in date:
                    date = datetime.now() - timedelta(hours=int(date.split(' ')[0]))
                elif 'min' in date:
                    date = datetime.now() - timedelta(minutes=int(date.split(' ')[0]))
                elif 'sec' in date:
                    date = datetime.now() - timedelta(seconds=int(date.split(' ')[0]))
                else:
                    date = datetime.now() - timedelta(days=int(date.split(' ')[0]))
            else:
                date = datetime.strptime(date, '%b %d, %Y')
            if date >= datetime.strptime(afterdate, "%Y%m%d"):
                date = date.strftime("%Y%m%d")
            else:
                continue
            title = validateTitle(episode.select('div[class="e3ZUqe"]')[0].text)
            fileUrl = episode.select('div[jsaction="rcuQ6b:vPK81"]')[0]['jsdata'].split(';')[1]
            print(title)
            resDownload = requests.get(fileUrl)
            if not os.path.exists('./podcast/{}/'.format(name)):
                os.makedirs('./podcast/{}/'.format(name))
            with open('./podcast/{}/{}.mp4'.format(name, title), 'wb') as f:
                f.write(resDownload.content)
            podcast.append({'Title' : title, 'Url': episodeurl, 'Date': date})
            print('Done')
            print('=='*30)

    with open(f'./podcast/{url_json}.json', 'w') as f:
        json.dump(podcast,f)
    return
```
