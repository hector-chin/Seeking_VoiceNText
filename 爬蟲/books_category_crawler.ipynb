```python
import requests
from bs4 import BeautifulSoup
import time
import json
import random
import os
```

```python
if not os.path.exists('./books'):
    os.mkdir('./books')
```

```python
url = 'https://www.books.com.tw/web/books/?loc=subject_main_001'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.54 Safari/537.36'
}
res = requests.get(url, headers=headers)

soup = BeautifulSoup(res.text, 'html.parser')
```

```python
for i, x in enumerate(soup.select('div[class="mod_b type02_l001-1 clearfix"]')[0].select('a')):
    if i in [1,3,4,5,7,8]:
        books = {}
        books[x.text] = {}
        print(x.text)
        url2 = x['href']
        res2 = requests.get(url2, headers=headers)
        soup2 = BeautifulSoup(res2.text, 'html.parser')
        time.sleep(random.randint(15,20))
        url3 = soup2.select('div[class="type02_bd"]')[0].a['href']
        res3 = requests.get(url3, headers=headers)
        soup3 = BeautifulSoup(res3.text, 'html.parser')
        time.sleep(random.randint(15,20))
        y = 0
        for j in soup3.select('li[class="item"]'):
            url4 = j.a['href']
            res4 = requests.get(url4, headers=headers)
            soup4 = BeautifulSoup(res4.text, 'html.parser')
            time.sleep(random.randint(15,20))
            try:
                se = soup4.select('div[class="grid_24 main_column"]')[0]
                ISBN = se.select('div[class="mod_b type02_m058 clearfix"]')[0].li.text.split('：')[1].strip()
                print(ISBN, end='...')
                books[x.text][ISBN] = {'title': se.h1.text,
                                       'content': se.select('div[class="content"]')[0].text.strip('\n'),
                                       'url': url4}
                print('Done')
            except IndexError as e:
                print(e)
                time.sleep(random.randint(10,15))
            y += 1
            if y%20 == 0:
                time.sleep(random.randint(55,65))
        with open('./books/books_{}.json'.format(x.text), 'w') as f:
            json.dump(books,f)
        print("=="*30)
        time.sleep(random.randint(45,65))
    elif i == 10:
        url2 = x['href']
        res2 = requests.get(url2, headers=headers)
        soup2 = BeautifulSoup(res2.text, 'html.parser')
        time.sleep(random.randint(15,20))
        for j in soup2.select('ul[class="sub"]')[0].select('a')[:3:2]:
            books = {}
            books[j.text] = {}
            print(j.text)
            url3 = j['href'].split('?')[0] + '?v=1&o=5'
            res3 = requests.get(url3, headers=headers)
            soup3 = BeautifulSoup(res3.text, 'html.parser')
            time.sleep(random.randint(15,20))
            y = 0
            for k in soup3.select('div[class="item"]'):
                url4 = k.a['href']
                res4 = requests.get(url4, headers=headers)
                soup4 = BeautifulSoup(res4.text, 'html.parser')
                time.sleep(random.randint(15,20))
                try:
                    se = soup4.select('div[class="grid_24 main_column"]')[0]
                    ISBN = se.select('div[class="mod_b type02_m058 clearfix"]')[0].li.text.split('：')[1].strip()
                    print(ISBN, end='...')
                    books[j.text][ISBN] = {'title': se.h1.text,
                                           'content': se.select('div[class="content"]')[0].text.strip('\n'),
                                           'url': url4}
                    print('Done')
                except IndexError as e:
                    print(e)
                    time.sleep(random.randint(10,15))
                y += 1
                if y%20 == 0:
                    time.sleep(random.randint(55,65))
            with open('./books/books_{}.json'.format(j.text), 'w') as f:
                json.dump(books,f)
            print('=='*30)
            time.sleep(random.randint(45,65))
    else:
        pass
```
